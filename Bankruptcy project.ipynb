```python
from scipy.io import arff
import pandas as pd
import numpy as np

pd.reset_option("display")
data = arff.loadarff('1year.arff')

df = pd.DataFrame(data[0])

df.head(10)
```

```python
df['class'].describe()
```

```python
df['class'].value_counts()
```

```python
# Substituting b'0' into 
df['class'].replace(b'0', '0',inplace=True) #Solvency
df['class'].replace(b'1', '1',inplace=True) #Bankrupt
pd.concat([df.head(), df.tail()])
```

```python
#Dealing with missing values
```

```python
print(df.isnull().sum())

```

```python
# Missing values percentage
missingV = df.isnull().sum()/df.notnull().sum()*100
print(missingV)
```

```python
# Attr21, 37 has 30% and 63% of missing values. Rest is less than 5%
# Apply mean or median replacement for Attr21, 37 and the rest we drop missing values
# Mean or Median?

print(df['Attr21'].describe(), df['Attr37'].describe())
```

```python
# Apply median, since we can find some outliers in Attr 21, 37 compared with Q1, Q2, and Q3

df['Attr21'] = df['Attr21'].fillna(df['Attr21'].median())
df['Attr37'] = df['Attr37'].fillna(df['Attr37'].median())
```

```python
print(df['Attr21'].describe(), df['Attr37'].describe())
```

```python
# Drop missing values for the rest variables
df1 = df.dropna(axis=0)
print(df1.isnull().sum())
```

```python
# Outlier detection
```

```python
# Dropping class variable for outlier detection
df2 = df1.iloc[:,:-1]
```

```python
#pd.concat([df2.head(), df2.tail()])
```

```python
def outliers(df2):
    Q1= df2.quantile(0.25)
    Q3 = df2.quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 20 * IQR
    lower_limit = Q1 - 20 * IQR
    return upper_limit, lower_limit

upper, lower = outliers(df2)
print("Upper whisker: ", upper)
print("Lower Whisker: ", lower)
```

```python
df2out = df2[(df2 > lower) & (df2 < upper)]
```

```python
#pd.concat([df2out.head(), df2out.tail()])
```

```python
df2out = df2out.join(df1["class"])
```

```python
#pd.concat([df2out.head(), df2out.tail()])
```

```python
df2out.isnull().sum()
```

```python
#Since it has humongous outlier numbers, I think it is good idea to use median
df2out = df2out.fillna(df.median())
```

```python
'''Applying classification models(Logistic regression, KNN, Random Forest, Gradient Boosting) to the datasets A1 to A5'''
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_validate
from sklearn.metrics import confusion_matrix

#A2 with Logistic regression
A2X = A2.iloc[:,0:18]  #Attributes
A2y = A2.iloc[:,-1]    #Class

scoring = ['accuracy', 'precision', 'recall' ,'roc_auc']

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter=2500)
cv = KFold(n_splits=5, random_state=1, shuffle=True)

A2y_p = cross_val_predict(clf, A2X, A2y, cv=cv)
conf_matA2 = confusion_matrix(A2y, A2y_p)

scoresA2LR = cross_validate(clf, A2X, A2y, scoring=scoring, cv=cv, n_jobs=-1)
scoresAcurracy = cross_validate(clf, A2X, A2y, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
# Requires normalization for better performance with KNN
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_A2KN = sc.fit_transform(A2X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
cv = KFold(n_splits=5, random_state=1, shuffle=True)
scoresA2KN = cross_val_score(classifier, X_A2KN, A2y, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
#A2 with Random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
cv = KFold(n_splits=5, random_state=1, shuffle=True)
scoresA2RF = cross_val_score(rf, A2X, A2y, scoring=['accuracy', 'precision', 'recall' ,'roc_auc'], cv=cv, n_jobs=-1)
```

```python
#A2 with Gradient Boosting Model
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
cv = KFold(n_splits=5, random_state=1, shuffle=True)
scoresA2GB = cross_val_score(gb, A2X, A2y, scoring=['accuracy', 'precision', 'recall' ,'roc_auc'], cv=cv, n_jobs=-1)
```

```python
#Comparing model scores with dataset A2
print(
    'Dataset A2',
    'Logistic Regression Accuracy: %.4f (%.4f)' % (mean(scoresA2LR), std(scoresA2LR)),
      'KNN Accuracy: %.4f (%.4f)' % (mean(scoresA2KN), std(scoresA2KN)), 
      'Random Forest Accuracy: %.4f (%.4f)' % (mean(scoresA2RF), std(scoresA2RF)), 
      'Gradient Boosting Accuracy: %.4f (%.4f)' % (mean(scoresA2GB), std(scoresA2GB)),
      sep='\n'
     )
```
